```python
class LinkedList:
  def __init__(self):
    self.head = None

  def is_empty(self):
    return self.head == None

  def get_tail(self):
    if self.is_empty():
      return None
    else:
      curr_node = self.head
      while curr_node.next != None:
        curr_node = curr_node.next
      return curr_node 

  def append(self, value):
    if self.head == None:
      self.head = Nodo(value)
    else: 
      curr_node = self.head
      while curr_node.next != None:
        curr_node = curr_node.next
      curr_node.next = Nodo(value)

  def prepend(self, value):
    if self.head == None:
      print('Está vacio, no se puede hacer prepend')
    else:
      aux = self.head
      self.head = Nodo(value, aux)

  def transversal(self):
    curr_node = self.head
    while curr_node != None:
      print(f'| {curr_node.data} |-->', end='')
      curr_node = curr_node.next
    print('')

  def remove(self, value):
    if self.head == None:
      print('Lista vacia no se puede remover nada')
    else:
      curr_node = self.head
      aux = None
      while curr_node.data != value:
        aux = curr_node
        curr_node = curr_node.next
      if curr_node != None:
        aux.next = curr_node.next
        curr_node = None

  def pop(self):
        if self.head == None :
            print("Lista vacía, no es posible regresar el ultimo valor.")
        else :
            curr_node=self.head
            aux = None
            while curr_node.next != None:
                aux= curr_node
                curr_node=curr_node.next
            aux.next = None
            return curr_node.data

  def pop_first(self):
        if self.head == None:
            print("Lista vacía, imposible eliminar el primer valor.")
        else:
            curr_node = self.head
            self.head = curr_node.next
            return curr_node.data
```

```python
lista = LinkedList()
if lista.is_empty():
  print('La lista está vacia')
else:
  print('La lista no está vacia')

lista.append(10)
lista.append(20)
lista.append(30)
lista.append(40)
lista.append(50)

print("Tail")
nodo = lista.get_tail()
if nodo != None:
  print(nodo.data)
else:
  print('La lista está vacia')

print("Transversal")
lista.transversal()

print("Remove")
lista.remove(20)
lista.transversal()

print("Prepend")
lista.prepend(0)
lista.transversal()

print("Pop")
print(f'Valor eliminado : {lista.pop()}')
lista.transversal()

print("Pop first")
print(f'Valor eliminado : {lista.pop_first()}')
lista.transversal()
```

#Listas Doblemente Ligadas

Estructura de datos ligada, que tiene doble nodo de puntero, que apuntan al siguiente y al anterior nodo.

- Apuntador que marca el punto de entrada(head)
- Apuntador que marca el final(tail) que tambièn es otro punto de entrada
- Permite el recorrido completo en ambas direcciones (transversal)



```python
class NodoDoble:
    def __init__(self, value, siguiente=None, previo=None):
        self.data = value
        self.next = siguiente
        self.prev = previo
```

```python
# ejercicio 1
print("Ejercicio 1")
head=NodoDoble(10)
tail=head
print(head.data)
print(tail.data)

# ejercicio 2
print("Ejercicio 2")
head=NodoDoble(10, NodoDoble(20))
head.next.prev=head
tail=head.next
print(f'El primero : {head.data}')
print(f'El primero : {tail.prev.data}')
print(f'El segundo : {head.next.data}')
print(f'El segundo : {tail.data}')

# ejercicio 3
print("Ejercicio 3")
head=NodoDoble(10, NodoDoble(20, NodoDoble(30, NodoDoble(40))))
tail=head.next.next.next
head.next.prev=head
head.next.next.prev=head.next
tail.prev=head.next.next
print(f'El primero : {head.data}')
print(f'El primero : {tail.prev.prev.prev.data}')
print(f'El ultimo: {tail.data}')
print(f'El ultimo : {head.next.next.next.data}')

# transversal
print("\nTransversal")
curr_node=head
while curr_node != None :
    print(f'| {curr_node.data} | -->', end="")
    curr_node=curr_node.next

# reverse transversal
print("\nReverse transversal")
curr_node=tail
while curr_node != None :
    print(f'| {curr_node.data} | -->', end="")
    curr_node=curr_node.prev

print("\nAgregar nodo al final")
tail.next=NodoDoble(50, None, tail)
tail=tail.next
curr_node=head
while curr_node != None :
    print(f'| {curr_node.data} | -->', end="")
    curr_node=curr_node.next
```

```python
class DoubleLinkedlist :
    def __init__(self) :
        self.head = None
        self.tail = None
        self.size = 0
 
    def get_size(self) :
        return self.size

    def is_empty(self) :
        return self.size == 0 

    def append(self, value) :
        if self.is_empty() :
            self.head=NodoDoble(value)
            self.tail = self.head
        else :
            self.tail.next = NodoDoble(value, None, self.tail)
            self.tail = self.tail.next
        self.size += 1

    def transversal(self) : 
        curr_node=self.head
        while curr_node != None :
            print(f'|{curr_node.data}| --> ', end="")
            curr_node = curr_node.next
        print("")    

    def find_from_tail(self, value) :
        curr_node = self.tail
        contador=1
        if self.is_empty() :
            print("No hay nodos.")
        else :
            while curr_node.data != value :
                curr_node = curr_node.prev
                contador+=1
            if curr_node.data == value:
                return contador

    def remove_from_tail(self, value) : 
        curr_node=self.tail
        while curr_node.data != value :
            curr_node = curr_node.prev
        if curr_node.data == value:
            curr_node.prev.next = curr_node.next
            curr_node.next.prev = curr_node.prev
        self.size -= 1

    def find_from_head(self, value) : 
        curr_node = self.head
        contador=1
        if self.is_empty() :
            print("No hay nodos.")
        else :
            while curr_node.data != value :
                curr_node = curr_node.next
                contador+=1
            if curr_node.data == value:
                return contador

    def remove_from_head(self, value) : 
        curr_node=self.head
        while curr_node.data != value :
            curr_node = curr_node.next
        if curr_node.data == value:
            curr_node.prev.next = curr_node.next
            curr_node.next.prev = curr_node.prev
        self.size -= 1
    
    def reverse_transversal(self) : 
        curr_node=self.tail
        while curr_node != None :
            print(f'|{curr_node.data}| --> ', end="")
            curr_node = curr_node.prev
        print("")
```

```python
ldd=DoubleLinkedlist()

ldd.append(10)
ldd.append(20)
ldd.append(30)
ldd.append(40)
ldd.append(50)

print("get size:")
print(f'{ldd.get_size()}')

print("is empty:")
print(f'{ldd.is_empty()}')

print("find from head:")
print(f'{ldd.find_from_head(30)}')

print("find from tail")
print(f'{ldd.find_from_tail(20)}')

ldd.remove_from_head(40)

ldd.remove_from_tail(20)

print("transversal")
ldd.transversal()

print("reverse_transversal")
ldd.reverse_transversal()
```
